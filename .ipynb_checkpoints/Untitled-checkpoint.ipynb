```python
with open('data', 'r') as f:
    data = f.readlines()
```

```python
drawn = data.pop(0)
```

```python
position = np.array(drawn.split(','), dtype=int)
position
```

```python
data = [line for line in data if line !='\n']
```

```python
data = [line.split(' ') for line in data]
```

```python
data = [[item for item in line if item !=''] for line in data]
```

```python
data = np.array(data, dtype=int)
```

```python
reshaped = data.reshape((-1, 5,5 ))
```

```python
def find_wining_grid(grid, positions):
    coordinates = []
    # go through possible number
    for number in positions:
        # write coordinate if number in grid
        results = np.where(grid == number)
        if len(results[0]) and len(results[1]):
            x,y = results[0][0], results[1][0]
        
            # get list of coordinates
            coordinates.append((x, y))
    
    # unique values in x or y
    coordo = pd.DataFrame(coordinates, columns=['x', 'y'])
    
    # if only one the grid is wining
    if any(coordo.x.value_counts()==5) or any(coordo.y.value_counts()==5):

    
        # return grid ?
        return grid
```

```python
find_wining_grid(reshaped[0,:,:], [78, 27, 82, 68, 7])
```

```python
reshape = [np.array(item) for item in reshaped.tolist()]
reshape
```

```python
def get_winning_values():
    for i in range(5, len(position)):
        posi = position[:i]
        for idx in range(len(reshaped)):
            grid = reshaped[idx, :, :]

            wining = find_wining_grid(grid, posi)

            if wining is not None:
                #import ipdb; ipdb.set_trace()
                return (wining, posi, posi[-1])
```

```python
a,b, c= get_winning_values()
```

```python
sum(set(a.flatten()).difference(b)) * c
```

```python
sum(set(a.flatten()).difference(b)) * c
```

```python
def get_last_winning(data):
    reshape = data.copy()
    for i in range(5, len(position)):
        posi = position[:i]
        idx_to_drop = []
        for idx, grid in enumerate(reshape):
         
            wining = find_wining_grid(grid, posi)

            if wining is not None:
                saved = wining
                saved_posi = posi
            else:
                idx_to_drop.append(idx)
                
    
        reshape = reshape.iloc[idx_to_drop]
                    
        
    return (saved, saved_posi, saved_posi[-1])
```

```python

f, g,h = get_last_winning(pd.Series(reshape))
```

```python
sum(set(f.flatten()).difference(g))*h
```

```python
sum(set(f.flatten()).difference(g))*42
```

```python
with open('day5' , 'r') as f:
    data = f.readlines()

data = pd.Series(data)
data
```

```python
def map_pattern(x):
    x = x.strip().replace(' -> ', ',')
    return x.split(',')
```

```python
data = data.str.strip()
data = data.str.split(' -> ')
```

```python
data = pd.DataFrame(data.to_list())
```

```python
data.iloc[:, 0] = data.iloc[:, 0].str.split(',')
data.iloc[:, 1] = data.iloc[:, 1].str.split(',')
```

```python
data.columns = ('init', 'final')
```

```python
data
```

```python
def convert_tup_int(x):
    return (int(x[0]), int(x[1]))

data = data.applymap(convert_tup_int)
data
```

```python
def create_axis_range(a, b):
    if a < b:
        return range(a, b+1)
    elif b < a:
        return range(a, b-1, -1)
    else:
        return None
        

def get_range(initial, final):
    if (initial[0] != final[0]) and (initial[1] == final[1]):
        x_range = create_axis_range(initial[0], final[0])
        res = [(x, initial[1]) for x in x_range]
        return res
    
    elif (initial[0] == final[0]) and (initial[1] != final[1]):
        y_range = create_axis_range(initial[1], final[1])
        res = [(initial[0], y) for y in y_range]
        return res
    
    else:
        return []
    
def get_positions(df):
    res = []
    for idx, row in df.iterrows():
        temp = get_range(row['init'], row['final'])
        res += temp
    return pd.Series(res)
```

```python
res = get_positions(data)
```

```python
print('result: ', (res.value_counts() >= 2).sum())
```

```python
def create_axis_range(a, b):
    if a < b:
        return range(a, b+1)
    elif b < a:
        return range(a, b-1, -1)
    else:
        return None
        

def get_range(initial, final):
    if (initial[0] != final[0]) and (initial[1] == final[1]):
        x_range = create_axis_range(initial[0], final[0])
        res = [(x, initial[1]) for x in x_range]
        return res
    
    elif (initial[0] == final[0]) and (initial[1] != final[1]):
        y_range = create_axis_range(initial[1], final[1])
        res = [(initial[0], y) for y in y_range]
        return res
    
    elif (initial[0] != final[0]) and (initial[1] != final[1]):
        x_range = create_axis_range(initial[0], final[0])
        y_range = create_axis_range(initial[1], final[1])
        

        res = [(x, y) for x, y in zip(x_range, y_range)]
        return res
    
    else:
        return []
    
    
def get_positions(df):
    res = []
    for idx, row in df.iterrows():
        temp = get_range(row['init'], row['final'])
        res += temp
    return pd.Series(res)
```

```python
res = get_positions(data)
res.value_counts().value_counts()
```

```python
print('result:', (res.value_counts() >=2).sum())
```
